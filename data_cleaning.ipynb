Separate Movies and TV Shows

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\User\\OneDrive\\Desktop\\NCA\\data\\netflix_cleaned.csv')
df.head()
```

```python
movies = df[df['type']=='Movie'].copy()

tvshows = df[df['type']=='TV Show'].copy()
```

```python
print(df.shape)
print(df.columns)
```

Extract Movie Duration

```python
movies['duration_min'] = movies['duration'].str.extract(r'(\d+)')
movies['duration_min'] = pd.to_numeric(
    movies['duration_min'],
    errors='coerce'
)
```

```python
movies['duration_min'].isna().sum()
```

```python
movies[movies['duration_min'].isna()][['title', 'duration']]
```

```python
movies = movies.dropna(subset=['duration_min'])
movies['duration_min'] = movies['duration_min'].astype(int)
```

```python
movies[['title','duration','duration_min']].head()
```

Extract Number of Seasons

```python
tvshows['seasons'] = tvshows['duration'].str.extract('(\d+)')

tvshows['seasons'] = tvshows['seasons'].astype(int)
```

```python
tvshows[['title','duration','seasons']].head()
```
